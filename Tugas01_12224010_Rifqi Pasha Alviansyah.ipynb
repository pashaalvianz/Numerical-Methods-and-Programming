## Metode Analitik

Persamaan kecepatan penerjun payung:

v(t) = (m g / c) (1 - e^(-(c/m)t))

Pada bagian ini dihitung solusi analitik untuk beberapa nilai koefisien hambatan:
c = 10, 12.5, dan 15 kg/s.


```python
# Analytical Solution 
 
# Fungsi untuk menghitung kecepatan jatuh parasut 
def hitung_v(massa, koef_hambatan, waktu): 
    g = 9.81  # Akselerasi gravitasi (m/s^2) 
    # Kecepatan Terminal (saat gaya setimbang) 
    kec_terminal = (massa * g) / koef_hambatan 
    # Kecepatan saat waktu t 
    kecepatan = kec_terminal * (1 - math.exp(- (koef_hambatan / massa) * waktu)) 
    return kecepatan 
 

#1 C = 10 
massa = 68.1  # massa penerjun payung (kg) 
koef_hambatan = 10 # koefisien hambatan (kg/s) 
waktu = 100  # waktu jatuh (detik) 
# Memanggil fungsi 
kecepatan = hitung_v(massa, koef_hambatan, waktu) 
# Cetak hasil 
print(f"Kecepatan setelah {waktu} detik: {kecepatan:.2f} m/s") 
 

#2 C = 12.5 
massa = 68.1  # massa penerjun payung (kg) 
koef_hambatan = 12.5 # koefisien hambatan (kg/s) 
waktu = 100  # waktu jatuh (detik) 
# Memanggil fungsi 
kecepatan = hitung_v(massa, koef_hambatan, waktu) 
# Cetak hasil 
print(f"Kecepatan setelah {waktu} detik: {kecepatan:.2f} m/s") 


#3 C = 15 
massa = 68.1  # massa penerjun payung (kg) 
koef_hambatan = 15 # koefisien hambatan (kg/s) 
waktu = 100  # waktu jatuh (detik) 
# Memanggil fungsi 
kecepatan = hitung_v(massa, koef_hambatan, waktu) 
# Cetak hasil 
print(f"Kecepatan setelah {waktu} detik: {kecepatan:.2f} m/s") 
```

## Metode Numerik

Metode Euler digunakan untuk menyelesaikan masalah ini:

dv/dt = g - (c/m)v

Perhitungan dilakukan untuk:
t = 0.1 s, 0.5 s, dan 1 s
serta c = 10, 12.5, dan 15 kg/s.


```python
#### Numerical Solution 

# Fungsi untuk menghitung kecepatan jatuh parasut
def hitung_v(massa, koef_hambatan, dt, t_maks): 
    g = 9.81  # gravitasi (m/s^2) 
    t = 0 
    v = 0 
    
    while t <= t_maks: 
        print(f"t = {t:.2f} s, v = {v:.4f} m/s") 
        v += (g - (koef_hambatan / massa) * v) * dt 
        t += dt 
 
# 1.  C = 10 dt = 0.1 
massa = 68.1  # kg 
koef_hambatan = 10  # kg/s 
dt = 0.1  # selang waktu (s) 
t_maks = 100  # total waktu (s) 
 
hitung_v(massa, koef_hambatan, dt, t_maks) 
 
# 2.  C = 10 dt = 0.5 
massa = 68.1  # kg 
koef_hambatan = 10  # kg/s 
dt = 0.5  # selang waktu (s) 
t_maks = 100  # total waktu (s) 
 
hitung_v(massa, koef_hambatan, dt, t_maks) 
 
# 3.  C = 10 dt = 1 
massa = 68.1  # kg 
koef_hambatan = 10  # kg/s 
dt = 1  # selang waktu (s) 
t_maks = 100  # total waktu (s) 
 
hitung_v(massa, koef_hambatan, dt, t_maks) 
 
# 4.  C = 12.5 dt = 0.1 
massa = 68.1  # kg 
koef_hambatan = 12.5  # kg/s 
dt = 0.1  # selang waktu (s) 
t_maks = 100  # total waktu (s) 
 
hitung_v(massa, koef_hambatan, dt, t_maks) 
 
# 5.  C = 12.5 dt = 0.5 
massa = 68.1  # kg 
koef_hambatan = 12.5  # kg/s 
dt = 0.5  # selang waktu (s) 
t_maks = 100  # total waktu (s) 
 
hitung_v(massa, koef_hambatan, dt, t_maks) 
 
# 6.  C = 12.5 dt = 1 
massa = 68.1  # kg 
koef_hambatan = 12.5  # kg/s 
dt = 1  # selang waktu (s) 
t_maks = 100  # total waktu (s) 
 
hitung_v(massa, koef_hambatan, dt, t_maks) 
 
# 7.  C = 15 dt = 0.1 
massa = 68.1  # kg 
koef_hambatan = 15  # kg/s 
dt = 0.1  # selang waktu (s) 
t_maks = 100  # total waktu (s) 
 
hitung_v(massa, koef_hambatan, dt, t_maks) 
 
# 8.  C = 15 dt = 0.5 
massa = 68.1  # kg 
koef_hambatan = 15  # kg/s 
dt = 0.5  # selang waktu (s) 
t_maks = 100  # total waktu (s) 
 
hitung_v(massa, koef_hambatan, dt, t_maks) 
 
# 9.  C = 15 dt = 1 
massa = 68.1  # kg 
koef_hambatan = 15  # kg/s 
dt = 1  # selang waktu (s) 
t_maks = 100  # total waktu (s) 
 
hitung_v(massa, koef_hambatan, dt, t_maks) 
```
